Geometric Transformation

```
import numpy as np

I = np.random.rand(4,4)

dims = I.shape

x = range(dims[1])
y = range(dims[0])

theta = -15/180 * np.pi

T = np.array([[np.cos(theta) , np.sin(theta), 0],
              [-np.sin(theta), np.cos(theta), 0],
              [0,              0,            1]])

Ig2 = np.zeros(dims)

for x1 in x:
  for y1 in y:
    coords = np.matmul(np.array([x1, y1, 1]), T).round().astype(int)
    # print(coords)
    if 0 <= coords[0] < dims[0] and 0 <= coords[1] < dims[1]:
      print('[{}, {}, 1] -> [{}, {}, 1]'.format(x1, y1, coords[0], coords[1]))
      Ig2[coords[0],coords[1]] = I[x1,y1]
print(I)
print(Ig2)
```
