Hough Circle Transform




```
import cv2
import scipy
import matplotlib.pyplot as plt
from IPython.display import Image, display
import numpy as np
from skimage.filters.thresholding import threshold_otsu
!wget 'https://github.com/Henry-Medeiros/images/raw/master/holes1.png'


img = cv2.imread('holes1.png',0)
img = cv2.cvtColor(img, cv2.COLOR_BAYER_BG2GRAY)

plt.imshow(img,'gray')
plt.show()
```

Hough 

```
I_edges = cv2.Canny(img,10,100)

plt.imshow(I_edges,'gray')
plt.show()
```

```

img_edges = cv2.Canny(img, 1, 100)

plt.imshow(img_edges, 'gray')
plt.show()

# ( image, method, dp( resolution), mindist, param1 , param2, *minRadius, *maxRadius )
circles = cv2.HoughCircles(img_edges, cv2.HOUGH_GRADIENT, 1, 10, param1 = 100, param2 = 30, minRadius =10 ,
                           maxRadius= 50) 

circles = np.uint16(np.around(circles))

for i in circles[0,:]:
  # an outer circle is drawn for visualization
  cv2.circle(img, (i[0],i[1]), i[2], (0,255,0),2)
  # its center is marked
  cv2.circle(img, (i[0],i[1]), 2,(0,0,255), 3)
  

plt.imshow(img)
plt.show()
```

Harris Corner Detector

```
import cv2
import scipy
import matplotlib.pyplot as plt
from IPython.display import Image, display
import numpy as np
!wget 'https://github.com/Henry-Medeiros/images/raw/master/corners.jpg'
img = cv2.imread('corners.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img,'gray')
plt.show()
```

```
from skimage.feature import corner_harris, corner_peaks, corner_subpix

#detect corner using corner Harris
corner_response = corner_harris(img, k = 0.1)

#detecting peak values
corner_peaks_val = corner_peaks(corner_response, 5 )



# corners_subpix_val = corner_subpix(img, corner_peaks_val, 5)
plt.imshow(corner_response, 'gray')
plt.show()

fig, ax = plt.subplots()
# Displaying the image.
ax.imshow(img, interpolation='nearest', cmap=plt.cm.gray)
x = corners_subpix_val[:, 1]
y = corners_subpix_val[:, 0]
ax.plot(x, y, 'ob', markersize=10)
# ax.axis('off')
# # Saving the image.

# plt.show()
```
